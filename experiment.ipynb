```python
import numpy as np
import matplotlib.pyplot as plt
```

RRM: Toy Example 

```python
def rrm_algorithm(theta_0, iterations):
    theta_values = [theta_0]
    for _ in range(iterations):
        theta_t = theta_values[-1]
        theta_next = 1 - theta_t
        theta_values.append(theta_next)
    return theta_values

theta_0 = 0.75  
iterations = 20
theta_sequence = rrm_algorithm(theta_0, iterations)
plt.figure(figsize=(12, 6))
plt.plot(theta_sequence, marker='o')
plt.title('Theta Values in RRM Algorithm Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Theta Value')
plt.xticks(range(iterations + 1))
plt.grid(True)
plt.show()

theta_sequence
```

Experimentation with eta for RRM:

```python
delta = 0.01
num_iter = []
def modified_rrm_algorithm(theta_0, iterations, eta):
    flag = False
    theta_values = [theta_0]
    for i in range(iterations):
        theta_t = theta_values[-1]

        theta_next = ((1 - theta_t) * 2 * eta + theta_t) / (1 + 2 * eta)
        if flag == False and theta_next < 0.5 + delta and theta_next > 0.5 - delta:
            num_iter.append(i)
            flag = True
        theta_values.append(theta_next)
    return theta_values


eta_values = [0.5, 1, 2, 5, 10, 20, 50, 100]
theta_0 = 0.75  
iterations = 500

results = {}
for eta in eta_values:
    results[eta] = modified_rrm_algorithm(theta_0, iterations, eta)

# Plotting the results
plt.figure(figsize=(12, 8))
for eta, theta_sequence in results.items():
    plt.plot(theta_sequence, marker='o', label=f'eta={eta}')
plt.title('Theta Values in Modified RRM Algorithm Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Theta Value')
plt.legend()
plt.xticks(range(0, iterations + 1, 10))
plt.grid(True)
plt.show()

results
```

```python
# Plotting the results
plt.figure(figsize=(12, 8))
plt.plot(eta_values, num_iter, marker='o')
plt.title('Number of Iterations for Modified RRM Convergence Under Error delta = 0.01')
plt.xlabel('eta')
plt.ylabel('Number of Iterations')
plt.legend()
# plt.xticks(range(0, iterations + 1, 10))
plt.grid(True)
plt.show()
```
